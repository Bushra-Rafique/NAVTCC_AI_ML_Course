```python
import pandas as pd

sample = 'Name,Age,City,Salary\nAli,25,Karachi,50000\nSara,30,Lahore,60000\nAhmed,22,Islamabad,45000\n'
with open('sample_data.csv', 'w') as f:
    f.write(sample)
```

```python
df = pd.read_csv('sample_data.csv')
df
```

```python
print('head(2):\n', df.head(2))
print('tail(1):\n', df.tail(1))
print('shape:', df.shape)
```

```python
df.to_csv('output.csv', index=False)
print('Saved to output.csv')
```
