# Task 48 – Pandas: Data Correlation
**Reference:** https://www.w3schools.com/python/pandas_correlations.asp

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'study_hours':[1,2,3,4,5,6,7,8],'score':[40,55,60,70,75,85,90,95],'sleep_hours':[8,7,7,6,6,5,5,4]}
df = pd.DataFrame(data)
df.corr()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df.plot.scatter(x='study_hours', y='score', ax=axes[0], title='Study vs Score')
df.plot.scatter(x='sleep_hours', y='score', ax=axes[1], title='Sleep vs Score', color='orange')
plt.tight_layout()
plt.show()
```
