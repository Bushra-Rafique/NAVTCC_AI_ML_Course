```python
def greet(name):
    """Simple greeting function."""
    print(f'Hello, {name}!')
```

```python
def add(a, b):
    """Returns the sum of two numbers."""
    return a + b
```

```python
def factorial(n):
    """Returns factorial of n (recursive)."""
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
def power(base, exp=2):
    """Returns base raised to exp (default exp=2)."""
    return base ** exp
```

```python
greet('World')
print('Sum:', add(3, 5))
print('5! =', factorial(5))
print('3^4 =', power(3, 4))
print('7 squared =', power(7))
```

Simple Statistics Function

```python
def get_division_details(dividend, divisor):
    """Returns multiple results: quotient and remainder."""
    quotient = dividend // divisor
    remainder = dividend % divisor
    return quotient, remainder

# 1. Calling with multiple return values
q, r = get_division_details(10, 3)
print(f"10 divided by 3: Quotient = {q}, Remainder = {r}")

# 2. Calling with keyword arguments
q2, r2 = get_division_details(divisor=5, dividend=22)
print(f"22 divided by 5: Quotient = {q2}, Remainder = {r2}")
```
