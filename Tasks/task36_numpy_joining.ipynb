```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print('concatenate:', np.concatenate((a, b)))
print('stack:\n', np.stack((a, b)))
print('hstack:', np.hstack((a, b)))
print('vstack:\n', np.vstack((a, b)))
```

```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])
print('concat axis=0 (rows):\n', np.concatenate((x,y), axis=0))
print('concat axis=1 (cols):\n', np.concatenate((x,y), axis=1))
```
