# Task 24 – Set Data Type in Python
**Reference:** https://www.programiz.com/python-programming/set

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print('A:', A)
print('B:', B)
```

```python
print('Union A|B:', A | B)
print('Intersection A&B:', A & B)
print('Difference A-B:', A - B)
print('Symmetric Diff A^B:', A ^ B)
```

```python
A.add(6)
A.discard(1)
print('After add(6), discard(1):', A)
print('3 in A:', 3 in A)
```

```python
nums = [1, 2, 2, 3, 4, 4, 5]
print('Remove duplicates:', sorted(set(nums)))
```
