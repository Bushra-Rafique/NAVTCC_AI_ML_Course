# Task 33 – NumPy: Array Shape
**Reference:** https://www.w3schools.com/python/numpy_array_shape.asp

```python
import numpy as np
print('1D shape:', np.array([1,2,3,4]).shape)
print('2D shape:', np.array([[1,2,3],[4,5,6]]).shape)
arr3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print('3D shape:', arr3.shape)
print('ndim:', arr3.ndim)
print('size:', arr3.size)
print('itemsize:', arr3.itemsize, 'bytes')
```
