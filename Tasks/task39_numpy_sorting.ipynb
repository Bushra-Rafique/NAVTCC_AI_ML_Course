# Task 39 – NumPy: Array Sorting
**Reference:** https://www.w3schools.com/python/numpy_array_sort.asp

```python
import numpy as np
arr = np.array([3,1,4,1,5,9,2,6])
print('Sorted asc:', np.sort(arr))
print('Sorted desc:', np.sort(arr)[::-1])
```

```python
arr2d = np.array([[3,1,2],[6,4,5]])
print('Sort by row:\n', np.sort(arr2d))
print('Sort by col:\n', np.sort(arr2d, axis=0))
```

```python
idx = np.argsort(arr)
print('argsort:', idx)
print('arr via argsort:', arr[idx])
print('Sorted strings:', np.sort(np.array(['banana','apple','cherry'])))
```
