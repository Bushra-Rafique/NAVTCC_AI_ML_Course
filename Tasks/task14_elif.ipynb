# Task 14 – If-Elif-Else Statement

```python
i = 20
if (i == 10):
    print('i is 10')
elif (i == 15):
    print('i is 15')
elif (i == 20):
    print('i is 20')
else:
    print('i is not present')
```

The Student Achievement Tracker

```python
score = float(input("Enter the student's score (0-100): "))

if score >= 90:
    print("Grade: A")
    print("Excellent work! You've mastered the material.")
elif score >= 80:
    print("Grade: B")
    print("Great job! You have a solid understanding.")
elif score >= 70:
    print("Grade: C")
    print("Good effort. You've passed, but there's room for improvement.")
elif score >= 60:
    print("Grade: D")
    print("You passed, but you should review the core concepts.")
else:
    print("Grade: F")
    print("You did not pass. Please schedule a meeting for extra help.")

print("\nReport generation complete.")
```
