```python
import pandas as pd, numpy as np
data = {'Name':['Ali','Sara',None,'Fatima','Usman'],'Age':[25,None,22,28,35],'Score':[88.5,92.0,78.3,None,70.4]}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
# Create a copy
df2 = df.copy()

# Fill missing values using direct assignment (The Best Practice)
df2['Age'] = df2['Age'].fillna(df2['Age'].mean())
df2['Score'] = df2['Score'].fillna(df2['Score'].median())
df2['Name'] = df2['Name'].fillna('Unknown')

# Display the result
df2
```
