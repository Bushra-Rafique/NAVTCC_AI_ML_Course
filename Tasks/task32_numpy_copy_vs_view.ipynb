# Task 32 – NumPy: Copy vs View
**Reference:** https://www.w3schools.com/python/numpy_copy_vs_view.asp

## View – shares memory

```python
import numpy as np
original = np.array([1,2,3,4,5])
view = original.view()
view[0] = 99
print('Original after view change:', original)
```

## Copy – independent

```python
original2 = np.array([1,2,3,4,5])
copy = original2.copy()
copy[0] = 99
print('Original2 after copy change:', original2)
print('Copy:', copy)
```

```python
print('view.base is original:', view.base is original)
print('copy.base is None:', copy.base is None)
```
