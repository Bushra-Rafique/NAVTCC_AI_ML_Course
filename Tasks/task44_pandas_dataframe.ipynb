```python
import pandas as pd
data = {'calories': [420, 380, 390], 'duration': [50, 40, 45]}
df = pd.DataFrame(data)
df
```

```python
df2 = pd.DataFrame(data, index=['day1', 'day2', 'day3'])
df2
```

```python
print('loc day2:\n', df2.loc['day2'])
print('iloc 0:\n', df2.iloc[0])
```
