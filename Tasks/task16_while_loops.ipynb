## Basic while loop

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

## while-else

```python
n = 0
while n < 3:
    print('n =', n)
    n += 1
else:
    print('Loop ended')
```

## Infinite loop with break

```python
i = 0
while True:
    print('i =', i)
    i += 1
    if i == 3:
        break
```

Secure Login Example

```python
correct_password = "python123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_input = input(f"Enter password (Attempt {attempts + 1}/{max_attempts}): ")
    
    if user_input == correct_password:
        print("Access Granted! Welcome to the system.")
        break  
    
    print("Incorrect password. Please try again.")
    attempts += 1
else:
    print("\nAccess Denied: Account locked. Too many failed attempts.")
```
