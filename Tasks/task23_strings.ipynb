**Reference:** https://www.programiz.com/python-programming/string

```python
s = 'Hello, Python!'
print('Length:', len(s))
print('Upper:', s.upper())
print('Lower:', s.lower())
print('Replace:', s.replace('Python', 'World'))
print('Split:', s.split(', '))
print('Strip:', '  hello  '.strip())
print('Find Python:', s.find('Python'))
```

```python
print('Slice [0:5]:', s[0:5])
print('Reversed:', s[::-1])
```

```python
name, age = 'Ali', 21
print(f'f-string: {name}, {age}')
print('format(): {} {}'.format(name, age))
print('%% operator: %s, %d' % (name, age))
```

```python
words = ['Python', 'is', 'fun']
sentence = ' '.join(words)
print('Joined:', sentence)
print('Split back:', sentence.split())
```
