# Task 50 – Standard Deviation
**Reference:** https://www.w3schools.com/python/python_ml_standard_deviation.asp

```python
import numpy as np
import matplotlib.pyplot as plt

data = [86, 87, 88, 86, 87, 85, 86]
mean = np.mean(data)
std  = np.std(data)
var  = np.var(data)
print('Mean:', mean)
print('Variance:', var)
print('Std Dev:', std)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data, 'o-', color='steelblue', label='Data')
plt.axhline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axhline(mean+std, color='orange', linestyle=':', label=f'+1σ: {mean+std:.2f}')
plt.axhline(mean-std, color='orange', linestyle=':', label=f'-1σ: {mean-std:.2f}')
plt.legend()
plt.title('Standard Deviation')
plt.tight_layout()
plt.show()
```
