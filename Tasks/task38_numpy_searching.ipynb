# Task 38 – NumPy: Array Searching
**Reference:** https://www.w3schools.com/python/numpy_array_search.asp

```python
import numpy as np
arr = np.array([1,2,3,4,5,4,3,2])
print('Indices where ==4:', np.where(arr == 4))
print('Conditional:', np.where(arr > 3, 'big', 'small'))
```

```python
s = np.array([1,3,5,7,9])
print('Insert 6 at:', np.searchsorted(s, 6))
```

```python
print('Index of max:', np.argmax(arr))
print('Index of min:', np.argmin(arr))
```
