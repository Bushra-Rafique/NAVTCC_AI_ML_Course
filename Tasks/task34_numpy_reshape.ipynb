# Task 34 – NumPy: Array Reshaping
**Reference:** https://www.w3schools.com/python/numpy_array_reshape.asp

```python
import numpy as np
arr = np.arange(1, 13)
print('Original:', arr)
```

```python
print('Reshape (3,4):\n', arr.reshape(3, 4))
print('Reshape (2,3,2):\n', arr.reshape(2, 3, 2))
```

```python
arr2d = np.array([[1,2,3],[4,5,6]])
print('flatten:', arr2d.flatten())
print('ravel:', arr2d.ravel())
print('Transpose:\n', arr2d.T)
```
