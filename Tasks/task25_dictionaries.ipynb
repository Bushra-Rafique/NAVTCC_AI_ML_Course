# Task 25 – Dictionary Data Type in Python
**Reference:** https://www.programiz.com/python-programming/dictionary

```python
student = {
    'name': 'Ali',
    'age': 21,
    'grade': 'A',
    'courses': ['Math', 'CS', 'Physics']
}
print(student)
print('Name:', student['name'])
print('Missing key:', student.get('email', 'Not found'))
```

```python
student['age'] = 22
student['email'] = 'ali@example.com'
del student['grade']
print('Updated:', student)
```

```python
for key, val in student.items():
    print(f'  {key}: {val}')
```

```python
squares = {x: x**2 for x in range(1, 6)}
print('Dict comprehension:', squares)
```
