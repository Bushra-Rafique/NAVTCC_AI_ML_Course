# Task 13 – Nested If Statement

```python
i = 10
if (i == 10):
    # First if statement
    if (i < 15):
        print('i is smaller than 15')
    # Nested if – only executes if statement above is true
    if (i < 12):
        print('i is smaller than 12 too')
    else:
        print('i is greater than 15')
```

The Tiered Discount System

```python
age = int(input("Enter your age: "))
print("Age Entered: ", age)

if age < 18:
    is_student = input("Are you a student? (yes/no): ").lower()
    
    if is_student == "yes":
        print("Ticket Price: $8 (Student Discount applied)")
    else:
        print("Ticket Price: $10 (Minor rate)")
        
else:
    if age >= 65:
        print("Ticket Price: $12 (Senior Discount)")
    else:
        print("Ticket Price: $15 (Standard Adult rate)")

print("Enjoy your movie!")
```
