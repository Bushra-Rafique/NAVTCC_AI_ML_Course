```python
import pandas as pd
a = [1, 7, 2]
s = pd.Series(a)
print(s)
```

```python
s2 = pd.Series(a, index=['x', 'y', 'z'])
print(s2)
print('y:', s2['y'])
```

```python
d = {'day1': 420, 'day2': 380, 'day3': 390}
s3 = pd.Series(d)
print(s3)
```
