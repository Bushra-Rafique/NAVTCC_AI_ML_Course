```python
import numpy as np
print('int:', np.array([1,2,3]).dtype)
print('float:', np.array([1.1,2.2]).dtype)
print('str:', np.array(['a','b']).dtype)
print('bool:', np.array([True, False]).dtype)
print('complex:', np.array([1+2j]).dtype)
```

```python
arr = np.array([1.5, 2.7, 3.9], dtype='i')
print('Float -> int:', arr)
```

```python
arr2 = np.array([1.1, 2.2, 3.3])
print('Cast to int:', arr2.astype(int))
print('Cast to complex:', arr2.astype(complex))
```
