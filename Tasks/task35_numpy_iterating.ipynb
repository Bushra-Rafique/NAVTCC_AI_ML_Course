# Task 35 – NumPy: Array Iteration
**Reference:** https://www.w3schools.com/python/numpy_array_iterating.asp

```python
import numpy as np
arr1 = np.array([1,2,3])
for x in arr1:
    print('1D:', x)
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
for row in arr2:
    for val in row:
        print('2D:', val)
```

```python
arr3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
for x in np.nditer(arr3):
    print('nditer:', x)
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
for idx, val in np.ndenumerate(arr2):
    print(idx, val)
```
