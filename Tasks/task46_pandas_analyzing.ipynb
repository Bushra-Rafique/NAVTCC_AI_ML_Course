```python
import pandas as pd
data = {'Name':['Ali','Sara','Ahmed','Fatima','Usman'],'Age':[25,30,22,28,35],'Score':[88.5,92.0,78.3,95.1,70.4]}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
print('Mean age:', df['Age'].mean())
print('Max score:', df['Score'].max())
print('Min score:', df['Score'].min())
```

```python
df[['Age','Score']].corr()
```
