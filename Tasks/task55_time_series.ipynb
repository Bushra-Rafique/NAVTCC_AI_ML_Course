```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
dates  = pd.date_range(start='2023-01-01', periods=365, freq='D')
values = 100 + np.cumsum(np.random.randn(365))
ts = pd.Series(values, index=dates)
ts.head()
```

```python
ts.describe()
```

```python
rolling_mean = ts.rolling(window=30).mean()
plt.figure(figsize=(12, 5))
plt.plot(ts, alpha=0.5, label='Original')
plt.plot(rolling_mean, color='red', label='30-Day Rolling Mean')
plt.title('Time Series with Rolling Mean')
plt.xlabel('Date'); plt.ylabel('Value')
plt.legend(); plt.tight_layout(); plt.show()
```

```python
monthly = ts.resample('ME').mean()
print('Monthly averages:\n', monthly.head())
print('\nFirst-order diff:\n', ts.diff().dropna().head())
```
