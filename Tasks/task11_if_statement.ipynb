```python
temperature = 35  # It is a hot day

if temperature > 30:
    # Because 35 > 30 is TRUE, this block WILL execute
    print('Warning: It is very hot outside! Remember to stay hydrated.')
    
print('Weather check complete.')
```

```python
purchase_amount = 150

if purchase_amount > 100:
    # This only runs if the amount is over 100
    print("You qualify for a 10% discount!")
    purchase_amount = purchase_amount * 0.9

print("Total amount to pay:", purchase_amount)
```
