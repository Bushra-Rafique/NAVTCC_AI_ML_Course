# Task 49 – Mean, Median and Mode
**Reference:** https://www.w3schools.com/python/python_ml_mean_median_mode.asp

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50, 20, 30, 20, 10]
print('Mean:', np.mean(data))
print('Median:', np.median(data))
mode_result = stats.mode(data, keepdims=True)
print('Mode:', mode_result.mode[0], '(count:', mode_result.count[0], ')')
```

```python
plt.figure(figsize=(8, 4))
plt.hist(data, bins=5, color='steelblue', edgecolor='black')
plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Mean: {np.mean(data)}')
plt.axvline(np.median(data), color='green', linestyle='--', label=f'Median: {np.median(data)}')
plt.legend()
plt.title('Mean, Median and Mode')
plt.tight_layout()
plt.show()
```
