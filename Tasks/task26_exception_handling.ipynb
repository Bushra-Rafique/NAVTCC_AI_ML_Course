## Basic try-except

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print('Error: Cannot divide by zero!')
```

## Multiple exceptions + finally

```python
def safe_convert(value):
    try:
        return int(value)
    except ValueError:
        print(f"ValueError: Cannot convert '{value}' to int")
    except TypeError:
        print('TypeError: Invalid type')
    finally:
        print('Attempt done.')

safe_convert('abc')
safe_convert(3.7)
```

## try-except-else

```python
try:
    num = int('123')
except ValueError:
    print('Invalid number')
else:
    print('Conversion successful:', num)
```

## Catch all exceptions

```python
try:
    lst = [1, 2, 3]
    print(lst[10])
except Exception as e:
    print(f'Exception: {type(e).__name__}: {e}')
```
