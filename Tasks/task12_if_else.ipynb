```python
i = 20
if (i < 15):
    print('i is smaller than 15')
    print("i'm in if Block")
else:
    print('i is greater than 15')
    print("i'm in else Block")
print("i'm not in if and not in else Block")
```

Login Status Check

```python
age = 14

if age >= 18:
    print("Access Granted: You are an adult.")
    print("Welcome to the dashboard.")
else:
    print("Access Denied: You must be 18 or older.")
    print("Redirecting to the homepage...")

print("Process finished.")
```

```python
#Take input from user
age = int(input("Enter your age: "))
print("Age entered: ", age)

if age >= 18:
    # This block runs if the condition is TRUE
    print("Access Granted: You are an adult.")
    print("Welcome to the dashboard.")
else:
    # This block runs if the condition is FALSE
    print("Access Denied: You must be 18 or older.")
    print("Redirecting to the homepage...")

# This line is not indented, so it ALWAYS runs
print("Process finished.")
```
