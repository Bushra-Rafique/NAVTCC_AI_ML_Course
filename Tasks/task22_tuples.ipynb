# Task 22 – Tuple Data Type in Python
**Reference:** https://www.programiz.com/python-programming/tuple

```python
coords = (10, 20, 30)
print('Tuple:', coords)
print('First:', coords[0])
print('Slice:', coords[0:2])
```

```python
x, y, z = coords
print(f'x={x}, y={y}, z={z}')
```

```python
single = (42,)
print('Single-element tuple:', single, type(single))
```

```python
t = (1, 2, 3, 2, 2, 4)
print('Count of 2:', t.count(2))
print('Index of 3:', t.index(3))
```
