## Basic range loop

```python
for i in range(5):
    print(i)
```

## Loop over a list

```python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
```

## enumerate()

```python
for index, fruit in enumerate(fruits):
    print(f'{index}: {fruit}')
```

## Nested for loop

```python
for i in range(3):
    for j in range(3):
        print(f'({i},{j})', end=' ')
    print()
```

## for-else

```python
for n in range(5):
    print(n)
else:
    print('Loop finished')
```

Classroom Seating Chart.

```python
# A list of students present in class
students = ['Alice', 'Bob', 'Charlie', 'David']

# We have 2 rows of desks, and each row has 2 seats
# This uses nested loops, range, enumerate, and for-else
print("--- Generating Seating Chart ---")

for row in range(2):  # Outer loop for rows
    print(f"Row {row + 1}:")
    
    for seat in range(2):  # Inner loop for Seats
        # Calculate a flat index to pull from the students list
        student_index = (row * 2) + seat
        
        if student_index < len(students):
            # Using enumerate style logic manually to show index and name
            current_student = students[student_index]
            print(f"  Seat {seat + 1}: {current_student}")
        else:
            print(f"  Seat {seat + 1}: [Empty]")
            
else:
    # The for-else block runs after the outer loop finishes naturally
    print("--- All rows have been processed ---")
```
