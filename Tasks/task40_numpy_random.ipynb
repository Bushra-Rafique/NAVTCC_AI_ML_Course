# Task 40 – NumPy: Random
**Reference:** https://www.w3schools.com/python/numpy_random.asp

```python
import numpy as np
print('Random float:', np.random.random())
print('Random int:', np.random.randint(1, 100))
print('Random (3,3):\n', np.random.rand(3, 3))
print('Normal dist:', np.random.randn(5))
```

```python
np.random.seed(42)
print('Seeded random:', np.random.rand(3))
```

```python
print('Choice:', np.random.choice([10,20,30,40,50], size=3))
arr = np.array([1,2,3,4,5])
np.random.shuffle(arr)
print('Shuffled:', arr)
```
