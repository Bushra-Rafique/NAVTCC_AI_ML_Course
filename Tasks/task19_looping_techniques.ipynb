## enumerate()

```python
fruits = ['apple', 'banana', 'cherry']
for idx, val in enumerate(fruits):
    print(idx, val)
```

## zip()

```python
names  = ['Ali', 'Sara', 'Ahmed']
scores = [85, 90, 78]
for name, score in zip(names, scores):
    print(f'{name}: {score}')
```

## items() on dict

```python
d = {'a': 1, 'b': 2, 'c': 3}
for key, val in d.items():
    print(key, '->', val)
```

## reversed() and sorted()

```python
for i in reversed(range(5)):
    print(i)

nums = [3, 1, 4, 1, 5, 9]
for n in sorted(nums):
    print(n)
```
