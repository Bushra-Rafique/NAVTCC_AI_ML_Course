# Task 37 – NumPy: Array Splitting
**Reference:** https://www.w3schools.com/python/numpy_array_split.asp

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])
print('Split into 3:', np.array_split(arr, 3))
print('Split at [2,4]:', np.array_split(arr, [2, 4]))
```

```python
arr2d = np.array([[1,2,3,4],[5,6,7,8]])
print('hsplit into 2:\n', np.hsplit(arr2d, 2))
print('vsplit into 2:\n', np.vsplit(arr2d, 2))
```
