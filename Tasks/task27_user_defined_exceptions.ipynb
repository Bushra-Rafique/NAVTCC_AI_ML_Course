# Task 27 – User Defined Exceptions in Python
**Reference:** https://www.programiz.com/python-programming/user-defined-exception

```python
class NegativeNumberError(Exception):
    def __init__(self, value):
        self.value = value
        super().__init__(f'Negative number not allowed: {value}')

class AgeOutOfRangeError(Exception):
    pass
```

```python
def square_root(num):
    if num < 0:
        raise NegativeNumberError(num)
    return num ** 0.5

def validate_age(age):
    if age < 0 or age > 120:
        raise AgeOutOfRangeError(f'Age {age} is out of range (0-120)')
    return age
```

```python
try:
    print(square_root(-9))
except NegativeNumberError as e:
    print(f'Caught: {e}')
```

```python
try:
    validate_age(150)
except AgeOutOfRangeError as e:
    print(f'Caught: {e}')
```

```python
print('sqrt(16):', square_root(16))
print('valid age:', validate_age(25))
```
